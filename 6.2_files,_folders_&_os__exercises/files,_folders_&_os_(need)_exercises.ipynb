# **Task 1**

```python
import os
import glob
import shutil
import numpy as np

month = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}

print(month.keys())
print(month.values())

path = 'oct-08_singapore.txt'
filename = path.split(os.path.sep)[-1]
date = filename.split('_')[0]
city = filename.split('_')[-1]

for key, values in month.items():
    date = date.replace(key, values)
print(f"{date}-{city}")
```

# **Task 2**

```python
def rename_my_file(old_file_name):
    path = 'oct-08_singapore.txt'
    filename = path.split(os.path.sep)[-1]
    date = filename.split('_')[0]
    city = filename.split('_')[-1]
    
    for key, values in month.items():
        date = date.replace(key, values)
    return f"{date}-{city}"

rename_my_file(old_file_name)
```

# **Task 3**

```python
oldfile = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']

for name in oldfile:
    def rename_my_file(old_file_name):
        path = name
        filename = path.split(os.path.sep)[-1]
        date = filename.split('_')[0]
        city = filename.split('_')[-1]
        
        for key, values in month.items():
            date = date.replace(key, values)
        print(f"{name} ----> {date}-{city}")
    rename_my_file(old_file_name)
```

# **Task 4**

```python
source = np.array(glob.glob('os*/**/*.txt', recursive=True))

for name in source:
    def rename_my_file(old_file_name):
        path = name
        filename = path.split(os.path.sep)[-1]
        date = filename.split('_')[0]
        city = filename.split('_')[-1]
        
        for key, values in month.items():
            date = date.replace(key, values)
        print(f"{name} ----> /.{date}-{city}")
    rename_my_file(old_file_name)
```
