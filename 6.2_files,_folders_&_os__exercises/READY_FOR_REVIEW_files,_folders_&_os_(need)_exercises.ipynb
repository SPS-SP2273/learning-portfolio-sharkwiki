# **Task 1**

```python
import os
import glob
import shutil

month_dict = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}

# print(month.keys())
# print(month.values())

path = 'oct-08_singapore.txt'
filename = path.split(os.path.sep)[-1]
date, city_ext = filename.split('_')
month, day = date.split('-')

print(f"{month_dict[month]}-{day}-{city_ext}")

# for key, values in month.items():
#     date = date.replace(key, values)
# print(f"{date}-{city}")
```

# **Task 2**

```python
def rename_my_file(path):
    try:
        filename = path.split(os.path.sep)[-1]
        date, city_ext = filename.split('_')
        month, day = date.split('-')
        return f"{month_dict[month]}-{day}-{city_ext}"
    except:
        return path

rename_my_file(path)
```

# **Task 4**

```python
source = glob.glob('os*/**/*.txt', recursive=True)

for old_file_name in source:
    testp = old_file_name
    prefix, filename = testp.split(os.path.sep)
   
    new_file_name = rename_my_file(filename)
    new_path = os.path.join(prefix, new_file_name)
    os.rename(old_file_name, new_path)

    
#   print(f"{filename} -> {new_file_name}")

# for name in source:
#     def rename_my_file(filename):
#         path = name
#         filename = path.split(os.path.sep)[-1]
#         date = filename.split('_')[0]
#         city = filename.split('_')[-1]
        
#         for key, values in month.items():
#             date = date.replace(key, values)
#             newname = f"{date}-{city}"
#             os.rename(filename, newname)
            
#     rename_my_file(filename)
```

# **Task 7**

```python
city = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila', 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

for x in city:
    old = glob.glob(f'os*/**/*{x}*', recursive=True)

    for file in old:
        try: 
            prefix, name = file.split(os.path.sep)
            old_path = os.path.join(prefix, name)
            new_path = os.path.join('city', x)
            shutil.move(old_path, new_path)
        except:
            print('File has already been moved.')
```
