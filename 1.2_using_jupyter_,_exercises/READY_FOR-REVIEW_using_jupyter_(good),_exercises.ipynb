## **Exercise 3 (Timing Stuff)**

```python
%%timeit
squares = [i**2 for i in range(10)] 
```

```python
%%timeit
squares = list()
for i in range(10):
    squares.append(i**2)
```

`%%timeit` ran the codeblock 7 * 1,000,000 times and output the mean and standard deviation of each codeblock. The first method is on average faster (1.34 s < 1.73 s) than the second method, because it does not have the extra step of also turning the output into a list.
